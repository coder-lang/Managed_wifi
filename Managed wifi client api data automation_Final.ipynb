# Operational clients data

```python
# mac_addresses list is passed here, working
# mac_addresses = ['80:BC:37:0C:E3:20', '80:BC:37:0C:07:A0']
api_url = "https://61.95.251.130:8443/wsg/api/public/v8_2/aps"
service_ticket = "ST-6919-onB0Qfk3l6gJRleU5O96-AirtelvSCG-Ruckus"

all_data_frames = []
duration = 1*180  # for 0.1 hour

start_time = datetime.now()
end_time = start_time + timedelta(seconds=duration)

while datetime.now() < end_time:
    for mac_address in mac_addresses:
        current_time = datetime.now()
        url_with_query_param = f"{api_url}/{mac_address}/operational/client?serviceTicket={service_ticket}"
        response = requests.get(url_with_query_param, verify=False)
        if response.status_code == 200:
            data = response.json()
            data['Timestamp'] = current_time
            data['MAC Address'] = mac_address  # Add the input MAC address to the output
            all_data_frames.append(pd.DataFrame(data))
            print(f"Data retrieved at {current_time} for MAC {mac_address}")
        else:
            print(f"Data retrieval failed with status code: {response.status_code} for MAC {mac_address}")

    time.sleep(180)  # gap of 3 min

print("Data retrieval completed.")

if all_data_frames:
    all_data = pd.concat(all_data_frames, ignore_index=True)
    print(all_data)
else:
    print("No data retrieved.")
```
